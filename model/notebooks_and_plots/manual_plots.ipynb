```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# cv scores
model_name_score = {
    'K-nearest Neighbors': [0.80786062, 0.93164104, 0.79413071, 0.66788474, 0.2658253],
    'Logistic Regression': [0.77983074, 0.88709516, 0.8850032, 0.72137761, 0.66928385],
    'Multi-layer Perceptron': [0.88363631, 0.99941165, 0.96024351, 0.98409161, 0.94562148],
    'Random Forest': [0.88363631, 0.99941165, 0.96024351, 0.98409161, 0.94562148],
}
```

```python
model_names = list(model_name_score.keys())
scores = list(np.mean(val) * 100 for val in model_name_score.values())
colors = plt.colormaps['Greys'](np.linspace(0.5, 1, len(scores)))

fig, ax = plt.subplots(figsize=(11, 4))
bars = ax.barh(model_names, scores, color=colors)

ax.set_xlabel('Score Percent')
ax.set_ylabel('Model')
ax.set_title('Model Test Score Comparison')

for i, bar in enumerate(bars):
    ax.text(x=bar.get_width() + 0.01, y=bar.get_y() + 0.4, s=f'{scores[i]:.2f}')

plt.savefig("model_test_score_comparison.svg", format="svg", bbox_inches='tight')
plt.show()
```

```python
model_name_training_time_seconds = {
    # 'Support Vector Machine': 39688.71,
    'Random Forest': 3615.3464,
    'Multi-layer Perceptron': 893.5561,
    'Logistic Regression': 547.6359,
    'K-nearest neighbors': 60.0630,
}

model_names = list(model_name_training_time_seconds.keys())
time_in_minutes = list(time_in_seconds/60 for time_in_seconds in model_name_training_time_seconds.values())
colors = plt.colormaps['viridis'](np.linspace(0.2, 0.8, len(time_in_minutes)))
fig, ax = plt.subplots(figsize=(15, 4.5))

bars = ax.bar(model_names, time_in_minutes, color=colors, log=False)
for bar in bars:
    x_pos = bar.get_x() + bar.get_width() / 2
    y_pos = bar.get_height()
    label = "{:.2f}".format(y_pos)
    ax.text(x_pos, y_pos, label, ha='center', va='bottom', fontsize=10)

ax.set_xlabel('Model')
ax.set_ylabel('Time Taken to Train (minutes)')
ax.set_title('Model Training Time Comparison')


plt.savefig("model_training_time_comparison.svg", format="svg", bbox_inches='tight')
plt.show()
```
