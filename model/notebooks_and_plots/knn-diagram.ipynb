```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
```

```python
neighbors = 3
X, y = make_blobs(n_samples=20, centers=2, random_state=0)
knn = KNeighborsClassifier(n_neighbors=neighbors)#, weights='distance')
knn.fit(X, y)

new_blob = np.array([[3, 3]])

fig, ax = plt.subplots()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
plt.scatter(new_blob[:, 0], new_blob[:, 1], color='green', marker='*', s=150)

ax.set_yticklabels([])
ax.set_xticklabels([])

distances, indices = knn.kneighbors(new_blob)
radius = distances[0, neighbors-1]
circle = plt.Circle(new_blob[0], radius+0.1, color='black', fill=False)
plt.gcf().gca().add_artist(circle)

plt.ylabel('Units')
plt.xlabel('Units')

plt.savefig("knn_example.svg")
plt.show()
```
